```python
from sklearn.preprocessing import LabelEncoder
```

```python
cate=[]
for i in dia.columns:
  if dia[i].dtypes==object:
    cate.append(i)
print(cate)
```

```python
label_Encoder = LabelEncoder()
for i in cate:
  dia[i]=dia[[i]].apply(label_Encoder.fit_transform)
```
